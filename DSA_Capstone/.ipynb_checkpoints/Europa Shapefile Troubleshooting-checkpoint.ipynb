The zoning data I received comes in a ZIP file with two SHP items: `prclz.shp` and `prclzm.shp`. It looks like this is the root of the issue. Simply pointing `geopandas.read_file()` at the `'data/zoning'` directory seems to default to pulling `prclzm.shp` when using the latest version of geopandas (on europa), whereas the older version (on GLAP) pulls `prclz.shp`, which is what I need. 

### Load the `prclz.shp` file

```python
# read in the zoning data we just unzipped
# zoning = gpd.read_file('data/zoning')
zoning = gpd.read_file('data/zoning/prclz.shp')
```

```python
# take a look the data
zoning.head()
```

```python
zoning.size
```

```python
# visualize the data
zoning.plot(figsize=(15,15), column='LAYER')
```

```python
zoning.LAYER.unique()
```

```python
zoning.loc[zoning['HANDLE'] == '10337000010']
```

```python
df = zoning.loc[zoning['HANDLE'] == '10337000010']
df.plot(column='LAYER')
```
